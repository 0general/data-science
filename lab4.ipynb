```python
# Lab4-3. 데이터셋 분할
setosadata = irisdf[irisdf['target']=='setosa']
setosadata
versicolordata = irisdf[irisdf['target']=='versicolor']
versicolordata
virginicadata = irisdf[irisdf['target']=='virginica']
virginicadata
```

```python
def iris_train_test_extract(X,Y,Z, test_size=0.25):
    
    test_num1 = int(Y.shape[0]*test_size)
    test_num2 = int(Y.shape[0]*test_size)
    test_num3 = int(Y.shape[0]*test_size)
    
    X_shuffled = X.sample(frac=1).reset_index(drop=True)
#     frac 매개변수 -> 반환해야 하는 전체 인스턴스 비율. 셔플하려면 1로 설정
    Y_shuffled = Y.sample(frac=1).reset_index(drop=True)
    Z_shuffled = Z.sample(frac=1).reset_index(drop=True)
#     rest_index 메소드 -> 데이터 프레임 인덱스 재설정
#     drop=True 옵션은 index 열이 새 열로 추가되지 않도록 한다.
    
    test_X = X_shuffled.iloc[:test_num1]
    train_X = X_shuffled.iloc[test_num1:]
    
    test_Y = Y_shuffled.iloc[:test_num2]
    train_Y = Y_shuffled.iloc[test_num2:]
    
    test_Z = Z_shuffled.iloc[:test_num3]
    train_Z = Z_shuffled.iloc[test_num3:]
    
    test_data = pandas.concat([test_X,test_Y,test_Z],ignore_index=True)
    train_data = pandas.concat([train_X,train_Y,train_Z], ignore_index=True)
    
    return test_data, train_data
```

```python
test_dataset, train_dataset = iris_train_test_extract(setosadata,versicolordata,virginicadata)
```

```python
test_dataset.to_csv('D:\\ds2021\\ds2021\\test_dataset.csv',index=False)
train_dataset.to_csv('train_dataset.csv',index=False)
```

```python
# 분할된 데이터셋 가시화
train_data = pandas.read_csv("train_dataset.csv", sep=',')
test_data = pandas.read_csv("test_dataset.csv", sep=',')

train_data.head()
test_data.head()
```

```python
trainmean = train_data.groupby(['target']).mean()
testmean = test_data.groupby(['target']).mean()
trainmean
testmean
```

```python
def make_bar(mdata, splt):
    
    label = ['SL', 'SW', 'PL', 'PW']
    index = np.arange(len(label))
    
    mallset = mdata.iloc[0]
    mallver = mdata.iloc[1]
    mallvir = mdata.iloc[2]
    
    splt.bar(index - 0.2, mallset, width = 0.2, label='setosa')
    splt.bar(index, mallver, width = 0.2,label='versicolor')
    splt.bar(index + 0.2,mallvir, width = 0.2, label='virginica')
    splt.legend(['setosa','versicolor','virginica'])
    
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.setp(axes, xticks = [0.0, 1.0, 2.0, 3.0], xticklabels=label)
make_bar(trainmean,axes[0])
make_bar(testmean,axes[1])
```


왼쪽은 train dataset, 오른쪽은 test dataset으로 그린 그래프이다. 거의 유사한 형태의 그래프 모양을 보여주고 있다.
