```python
import numpy as np
import matplotlib.pyplot as plt
```

미분 정의를 이용하는 간단한 미분계산

```python
f1 = (lambda x : x**2)
f2 = (lambda x : x**3 + x**2 + x)
f3 = (lambda x : x*np.sin(x*x*x+3))
f4 = (lambda x : np.cos(np.sqrt(np.exp(x)+1)/2))

def der_fun(f, a):
    h = 0.0000000001
    return (f(a+h)-f(a))/h
  
print('f1\'(2) = ', der_fun(f1, 2))
print('f2\'(2) = ', der_fun(f2, 2))
print('f3\'(2) = ', der_fun(f3, 2))
print('f4\'(2) = ', der_fun(f4, 2))
```

```python
x = np.linspace(-10,10,100)

fig, ax = plt.subplots(nrows=2, ncols=2)


ax[0,0].plot(x,f1(x))
ax[0,0].plot(x,der_fun(f1,x))

ax[0,1].plot(x,f2(x))
ax[0,1].plot(x,der_fun(f2,x))

ax[1,0].plot(x,f3(x))
ax[1,0].plot(x,der_fun(f3,x))

ax[1,1].plot(x,f4(x))
ax[1,1].plot(x,der_fun(f4,x))

```

 
 각 그래프의 모형을 따로따로 확인해보자.
 

```python
plt.plot(x,f1(x))
```

 위의 그래프는 f1(x)이고 아래 함수는 f'1(x)이다.
 우리가 잘 아는 이차함수와 일차함수의 형태다.

```python
plt.plot(x,der_fun(f1,x))
```

```python
plt.plot(x,f2(x))
```

 위의 그래프는 f2(x)이고 아래 함수는 f'2(x)이다.
 f2(x)는 3차함수, f2'(x)는 2차함수의 모형이다.

```python
plt.plot(x,der_fun(f2,x))
```

```python
plt.plot(x,f3(x))
```

 위의 그래프는 f3(x)이고 아래 함수는 f'3(x)이다.
 작게 보았을 때는 f3(x)그래프가 가로 선형 그래프처럼 보였으나,
 실제로는 나비모양으로 진동하는 것을 알 수 있다. 
 함수와 도함수 값의 크기 차이가 상당하여 작게 그렸을 때는 잘 보이지 않았던 모양을 확인할 수 있다.

```python
plt.plot(x,der_fun(f3,x))
```

```python
plt.plot(x,f4(x))
```

 위의 그래프는 f4(x)이고 아래 함수는 f'4(x)이다.
 마찬가지로 함수와 도함수간의 값의 크기 차이가 커서 작게 보았을 때는 확인할 수 없었던 그래프의 모양을 자세히 볼 수 있다.

```python
plt.plot(x,der_fun(f4,x))
```
